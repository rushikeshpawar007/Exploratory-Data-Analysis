```python
df['Salary Estimate'].unique()
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K',''))
```

```python
df['Salary Estimate']
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.lower().replace('employer provided salary','').replace('per hour',''))
df['Salary Estimate'] = df['Salary Estimate'].apply(lambda x: x.lower().replace(':',''))
```

```python
df['Salary Estimate'].unique()
```

```python
df['min_salary2'] = df['Salary Estimate'].apply(lambda x: x.split('-')[0])
```

```python
df['max_salary2'] = df['Salary Estimate'].apply(lambda x: x.split('-')[1])
```

```python
df['avg_salary2'] = df['min_salary2'] + df['max_salary2']
```

```python
df['avg_salary2'] = df['avg_salary2'].astype(int)
```

```python
df['avg_salary2'] = df['avg_salary2'] / 2
```

```python
df.drop(['min_salary', 'max_salary', 'hourly', 'avg_salary'], axis=1)
df
```
