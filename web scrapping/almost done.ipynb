```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
```

```python
browser = webdriver.Chrome('chromedriver.exe')
browser.get('https://www.linkedin.com/uas/login')
file = open('config.txt')
lines = file.readlines()
username = lines[0]
password = lines[1]


elementID = browser.find_element_by_id('username')
elementID.send_keys(username)

elementID = browser.find_element_by_id('password')
elementID.send_keys(password)

elementID.submit()
```

```python
fulllink = 'https://www.linkedin.com/my-items/saved-jobs/?start=10'
browser.get(fulllink)
```

```python
visitedprofile = []
profilesqueued = []
```

```python
def getnewprofileID(soup, profilesqueued):
    profileID = []
    mj = soup.find('section', {'class' : 'grid__col grid__col--lg-12 mb6'})
    all_links = mj.findAll('a', {'class':'app-aware-link'})
    for link in all_links:
        userID = link.text
        if((userID not in profilesqueued) and (userID not in visitedprofile)):
            profileID.append(userID)
    return profileID
```

```python
new = getnewprofileID(BeautifulSoup(browser.page_source), profilesqueued)
```
