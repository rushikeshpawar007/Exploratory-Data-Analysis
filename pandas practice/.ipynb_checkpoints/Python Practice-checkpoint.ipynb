```python
nba2 = pd.read_csv('nba.csv').dropna(how = 'all')
nba2['Salary'] = nba2['Salary'].fillna(0).astype('int')
nba2.tail(3)
```

```python
nba2['Salary Rank'] = nba2['Salary'].rank(ascending=False).astype('int')
```

```python
nba2.head(3)
```

```python
nba2['Salary Rank'].sort_values(ascending=True)
```

```python
nba2[nba2['Salary Rank'] == 1]
```

```python
nba2.sort_values(by = 'Salary', ascending= False)
```
